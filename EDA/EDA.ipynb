# Data for Model 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### CO2 emissions data

```python
co2_data = pd.read_csv('../clean_data/annual-co-emissions-from-aviation_eu.csv')
print(co2_data.head())
#emissions in tonnes
```

```python
print(co2_data.describe()) 
```

```python
sns.set(style='whitegrid', context='talk')

pivot = co2_data.pivot(index='Year', columns='Entity', values='Total annual CO₂ emissions from aviation')
pivot.plot(figsize=(12,6), linewidth=2)
plt.ylabel('CO2 emissions (tonnes)')
plt.title('Top emitters — time series')
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.tight_layout()
plt.show()
```
